6- Write a Python Script to print even length words in a string-

Input: s = "I am Shailja"

Output: am


```python
s = "I am Shailja"
a = []
b = ""
for i in s:
  if(i == " "):
    a.append(b)
    b = ""
    continue
  else:
    b += i
a.append(b)    
for i in range(len(a)):
  if(i % 2 != 0):
    print(a[i],end = " ")
```

b: Find the data type and change it to float


```python
import numpy as np
Raw_data = [1,2,3,4,5,1,2,1,1,1]
arr = np.array(Raw_data)
count = {}
for i in range(arr.size):
  if(arr[i] in  count.keys()):
   count[arr[i]] += 1
  else:
   count[arr[i]] = 1
most_freq_value = count[max(count.values())]
print(most_freq_value)
arr = arr.astype('float64')
print(arr.dtype)
print(arr)
```

3- Create a Pandas Data frame from the given data -
raw_Data = {
              
              'Date':['10/2/2011', '11/2/2011', '12/2/2011', '13/2/2011'],
           'Product':['Umbrella', 'Matress', 'Badminton','Shuttle'],
             'Price':[1250, 1450, 1550, 400],
           'Expense': [ 21525220.653, 31125840.875, 23135428.768, 56245263.942]}

a- Add Index as Item1, Item2, Item3, Item4 

b- Find the index labels of all items whose ‘Price’ is greater than 1000.

c- Replace products using Map() with respective codes- Umbrella : ‘U’, Matress : 'M', Badminton : 'B', Shuttle: 'S'

d- Round off the Expense column values to two decimal places.

e- Create a new column called ‘Discounted_Price’ after applying a 10% discount on the existing ‘price’ column.(try using lambda function)

f- Convert the column type of “Date” to datetime format 

g- Create a column rank which ranks the products based on the price (one with highest price will be rank 1).


```python
import pandas as pd
df = pd.DataFrame({ 'Date':['10/2/2011', '11/2/2011', '12/2/2011', '13/2/2011'],
                 'Product':['Umbrella', 'Matress', 'Badminton','Shuttle'],
                   'Price':[1250, 1450, 1550, 400],
                 'Expense': [21525220.653, 31125840.875, 23135428.768, 56245263.942]})
df.columns.values[0] = "item1"
df.columns.values[1] = "item2"
df.columns.values[2] = "item3"
df.columns.values[3] = "item4"
#print(df)
# b)
df1 =df[df['item3'] > 1000] 
print(df1)
# c)
df['item2'] = df['item2'].map({'Umbrella':'U',
                             'Matress':'M',
                             'Badminton':'B',
                             'Shuttle':'S'},
                             na_action=None)
#print(df)
# d)
df['item4'] = df['item4'].round(decimals = 2)
print(df)
# e)
df['Discounted_Price'] = list(map(lambda x : x * 0.9,df['item3'].values))
print(df)
# f)
df['item1']= pd.to_datetime(df['item1'])
print(df)
# g)
df["Rank"] = df["item3"].rank(ascending = False)
df.sort_values("Rank", inplace = True,)
print(df)
```
